```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import random
%matplotlib inline

import chainer
import chainer.functions as F
import chainer.links as L
from chainer import report, training, Chain, datasets, iterators, optimizers
from chainer.training import extensions
from chainer.datasets import tuple_dataset
from chainer import Variable

from sklearn.model_selection import train_test_split
```

```python
min_temp_raw = []
with open("../data/daily-minimum-temperatures-in-me.csv") as f:
    reader = csv.reader(f)
    for row in reader:
        min_temp_raw.append(row)
#print(min_temp_raw)
header = min_temp_raw[0]
min_temp_raw = np.array(min_temp_raw[1:-3])
```

```python
min_temp_df = pd.DataFrame([e[1] for e in min_temp_raw],
                           columns = ["temperature"],
                           index = [e[0] for e in min_temp_raw])

#?0.2のような欠陥データを修正
for date in min_temp_df.index:
    t = min_temp_df.loc[date][0]
    try:
        min_temp_df.loc[date] = float(t)
    except:
        t = t.replace("?", "-")
        print(t)
        min_temp_df.loc[date] = float(t.replace("?", "-"))
```

```python
temperature = min_temp_df["temperature"].values
plt.plot(temperature)
```

# 時系列データの前処理:正規化と移動平均

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(temperature.reshape(-1,1))
temperature_minmax = scaler.transform(temperature.reshape(-1,1)).reshape(-1)
plt.plot(temperature_minmax)
```

```python
moving_ave=np.ones(60)/60

temperature_ave = np.convolve(temperature_minmax, moving_ave, mode='same')#移動平均

#周期構造を持つデータのみ取得
temperature_ave = temperature_ave[30:-30]

date = range(0, len(temperature_ave))
plt.plot(date,temperature_ave)
plt.show()
```

```python
#これを、ノイズはそのままに589ステップずつに区切って、合計3284個のデータを作成する。
n_step = 589
X, y = [],[]
for i in range(len(temperature_ave)-n_step-1):
    X.append(temperature_ave[i:i+n_step].astype(np.float32))
    y.append([temperature_ave[i+n_step]])
    
print(len(X))
print(len(y))
```

# chainer でLSTMを実装しよう

```python
import chainer
import chainer.functions as F
import chainer.links as L
```

```python
#365日間の気温推移から、366日目の気温を予測する
lstm = L.LSTM(None, 1)
print(lstm(np.array([X[0]])).shape)
```

```python
#どのような構成にするかは自由。
#シンプルなもので実装。

class LSTM(Chain):
    
    def __init__(self, n_out, n_units):
        super(LSTM, self).__init__(
            linear1 = L.Linear(None, n_units),
            lstm = L.LSTM(None, n_units),
            linear2 = L.Linear(None, n_out),
        )

    def reset_state(self):
        self.lstm.reset_state()
        self.zerograds()

    def __call__(self, x, t=None, train=False):
        
        x = Variable(x)
        if train:
            t = Variable(t)
        h1 = self.linear1(x)
        h2 = self.lstm(h1)
        y = self.linear2(h2)
        
        if train:
            return F.mean_squared_error(y[0], t)
        else:
            return y.data
```

```python
n_epoch = 20
n_data_in_batch = 20

model = LSTM(1, 10)
optimizer = optimizers.Adam(alpha=0.001)
optimizer.setup(model)

for epoch in range(n_epoch):

    index_random = list(range(0,len(X_train)))
    random.shuffle(index_random)
    total_loss = 0
    for i in range(0, len(X_train)-n_data_in_batch+1, n_data_in_batch):
        X_batch, y_batch = [], []
        index_list = index_random[i:i+n_data_in_batch]
        for index in index_list:
            X_batch.append([X_train[index]])
            y_batch.append(y_train[index])
        X_batch = np.array(X_batch, dtype="float32")
        y_batch = np.array(y_batch, dtype="float32")

        train_loss = 0
        model.reset_state()
        
        for X_, y_ in zip(X_batch, y_batch):
            train_loss += model(X_, t=y_, train=True)

        train_loss.backward()
        train_loss.unchain_backward()
        total_loss += train_loss.data/len(X_batch)
        optimizer.update()
    total_loss /= int(len(X_train)/n_data_in_batch)
    

    test_loss = 0
    for X_, y_ in zip(X_test, y_test):
        test_loss += model(np.array([X_], dtype="float32"), np.array(y_, dtype="float32"), train=True).data/len(X_test)
    print("epoch:\t{}\ttotal loss:\t{}, test loss:\t{}".format(epoch+1, total_loss, test_loss))
    predicted_temperature = []
    initial_temperature = np.array(temperature_ave[:n_step], dtype="float32") # 予測直前までの時系列
    N=len(temperature_ave)

    for _ in range(0, N-n_step):
        model.reset_state()
        y = model(np.array([initial_temperature], dtype="float32"), train=False)
        predicted_temperature.append(y[0])
        #予測された天気を更新。これをもとに次の予測をする。
        initial_temperature = np.delete(initial_temperature,0)
        initial_temperature = np.append(initial_temperature,y[0])

    plt.plot(range(0, n_step), temperature_ave[:n_step], color="green", label="temperature")
    plt.plot(range(n_step, N), temperature_ave[n_step:], color="red", label="temperature")
    plt.plot(range(n_step, N), predicted_temperature, color="blue", label="predicted")

    #plt.savefig("adam_alpha0.001_hidden3_nstep589_epoch" + str(epoch) + ".png")
    plt.show()
```
